# Example of the EPICS Archiver Interface

This example illustartes an interface to a hybrid data store consisting of data catalog and data files. Currently, it is implemented around the original version of the Channel Archiver file system consisting of index and data files. 

The interface is designed after the Spark programming model and MongoDB API and extends them for hybrid systems. As a result, the interface consists of the following steps:
1. create proxy associated with a collection of data catalog
2. aggregate queries (a la Spark transformations) on the client layer
3. apply the Spark action (e.g. collect) by transferring it together
   with accumulated transformations to the middle layer

First, we import the necessary modules from the python standard library, numpy, pandas, matplotlib, and eapy.

```python
#import the standard modules
import sys
import pdb
import time
from datetime import timedelta, datetime, tzinfo

#import numpy, pandas, matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
# make graphics inline
%matplotlib inline

#import the eapy module
sys.path.append('/opt/eapy')
import eapy
```

and define parameters of the EPICS V4 RPC service, such as a channel name and timeout.

```python
chName = "ea4.rpc:classic"; # service based on the original index files
# chName = "ea4.rpc:mongo"; # service based on MongoDB
timeout = 20.0;
```

Then, we need to start a session for connecting it with the middle layer service and creating a session-specific context.

```python
session = eapy.Session.createSession();
session.open(chName, timeout);
```

Now we can apply the Spark-oriented interface for accessing our hybrid data sytem. In the EPICS Archiver, data are partitioned into multiple subarchives associated with corresponding control subsystems. In this example, we use records of EPICS process variables (pvs) acquired from the RadMon (radiation monitor) subsystem.

```python
pvsProxy = session.createProxy("RadMon_2015");
```

According to our approach, a single request aggregates two queries. The first query (filter) is associated with a data catalog and we need to specify a pattern of selected pvs. (Note: The original archiver service does not support regular expressions with '/', '{', '}' and ignores the $project operator.)

```python
# pvsFilter = pvsProxy.filter([{"$match" : {"pv" : {"$regex" : "BR-AM.*RadMon:01.*Itvl:Data-I"}}},
pvsFilter = pvsProxy.filter([{"$match" : {"pv" : {"$regex" : ".*Itvl:Data-I"}}},
# pvsFilter = pvsProxy.filter([{"$match" : {"pv" : {"$regex" : ".*"}}},
                             {"$project" : {"pv" : 1}}]);
```

The second query (read) is associated with data files and contains a time window (e.g. one day).

```python
# datetime(year, month, day, hour, minute, sec, usec)
end   = datetime(2015, 3, 11, 14, 58, 21, 37799);
start = end + timedelta(days = -1);

pvsRead = pvsFilter.read([{"$match" : {"start" : start , "end" : end}} ]);
```

Now we can apply an action to collect data sets defined in aggregated queries. 

```python
t1 = datetime.now();
pvsCollect = pvsRead.collect(timeout);
t2 = datetime.now()

print "processing time: ", (t2 - t1);
```

The collect method returns a structure including a query status and a collection of documents. First, let's check the query status (int type, string message, string callTree).

```python
print pvsCollect['status']['type']
print pvsCollect['status']['callTree']
```

There are four types of query statuses: 0 - OK, 1 - Warning, 2 - Error, 3 - Fatal. If the query is OK, we can extract an array of documents. 

```python
pvs = pvsCollect['docs']
print pvs['size']
```

According to the MongoDB-oriented model, any collections are heterogeneous. As a result, it is represented by the dynamic structure with size and index fields, like 0, 1, etc. First, let's plot values of the selected pv (for example, 4).

```python
dbrs       = pvs['4']['dbrs'];
values     = dbrs.ndarray('values');
datetimes  = dbrs.datetimes('secs', 'nanos');

# plot
plt.plot(datetimes, values)
# beautify the x-labels
plt.gcf().autofmt_xdate()
plt.show()
```

Now we can prepare a numpy homogeneous structured array with metadata (e.g. name, type, count, size) of all pvs 

```python
# structured array: name, type, count, size
size = pvs['size'];
pvsNP = np.zeros((size,),dtype=[('pv name', 'a50'),('type', 'i4'),('vs/record', 'i4'),('records', 'i4')])
for i in range(size):
    pv = pvs[str(i)]
    pvsNP[i] = (pv['name'], pv['type'], pv['count'], pv['size'])
```

and print it as the corresponding pandas-based data frame. The type (10) corresponds to double. 

```python
pvsDF = pd.DataFrame(pvsNP);
pvsDF
```

Finally, we close this session ...

```python
session.close();
```
