# Example of the EPICS Archiver Interface (Snapshot case)

This example illustartes an interface to a hybrid data store consisting of data catalog and data files. Currently, it is implemented around the original version of the Channel Archiver file system consisting of index and data files. 

The interface is designed after the Spark programming model and MongoDB API and extends them for hybrid systems. As a result, the interface consists of the following steps:
1. create proxy associated with a collection of data catalog
2. aggregate queries (a la Spark transformations) on the client layer
3. apply the Spark action (e.g. collect) by transferring it together
   with accumulated transformations to the middle layer

First, we import the necessary modules from the python standard library, numpy, scipy, pandas, matplotlib, and eapy.

```python
#import the standard modules
import sys
import pdb
import time
from datetime import timedelta, datetime, tzinfo
import math

#import numpy, scipy, pandas, matplotlib
import numpy as np
from scipy import ndimage
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
# make graphics inline
%matplotlib inline

#import the eapy module
sys.path.append('/opt/eapy')
import eapy
```

and define parameters of the EPICS V4 RPC service, such as a channel name and timeout.

```python
chName = "ea4.rpc:classic";
timeout = 3.0;
```

Then, we need to start a session for connecting it with the middle layer service and creating a session-specific context.

```python
session = eapy.Session.createSession();
session.open(chName, timeout);
```

Now we can apply the Spark-oriented interface for accessing our hybrid data sytem. In the EPICS Archiver, data are partitioned into multiple subarchives associated with corresponding control subsystems. In this example, we use records of EPICS process variables (pvs) acquired from the RadMon (radiation monitor) subsystem. 

```python
pvsProxy = session.createProxy("RadMon/2015");
```

According to our approach, a single request aggregates two queries. The first query (match) is associated with a data catalog and we need to specify an archive name and pattern of selected pvs. 

```python
pvsMatch = pvsProxy.aggregate({"$match" : {"name" : ".*"}});
```

The second query (read) is associated with data files and contains a time window and number of records (e.g. 1 for snapshot).

```python
# datetime(year, month, day, hour, minute, sec, usec)
end   = datetime(2015, 3, 11, 14, 58, 21, 37799);
start1 = end + timedelta(days = -1);

pvsRead1 = pvsMatch.aggregate({"$read" : {"start" : start1 , "end" : end, "limit" : 1}});
```

Now we can apply an action to collect data sets defined in aggregated queries. 

```python
t1 = datetime.now();
pvsDoc1 = pvsRead1.collect(timeout);
t2 = datetime.now()

print "processing time: ", (t2 - t1);
```

Let's check another snapshot

```python
start2 = end + timedelta(days = -10);
pvsRead2 = pvsMatch.aggregate({"$read" : {"start" : start2 , "end" : end, "limit" : 1}});
t1 = datetime.now();
pvsDoc2 = pvsRead2.collect(timeout);
t2 = datetime.now()

print "processing time: ", (t2 - t1);
```

The $collect$ action returns a heterogeneous structured array of pv-specific documents (name, meta, type, count, size, and pv-specific dbrs). Since EPICS V4 (as well as numpy) uses homegeneous collections, this array is represented by a dynamic structure with $size$ and index fields, like $0$, $1$, etc. 

```python
size = min(pvsDoc1['size'], pvsDoc2['size']);
print size;
```

Now we can plot them side-by-side

```python
size1 = math.floor(math.sqrt(size));
size2 = math.trunc(size1*size1);

valuesNP1 = np.zeros((size2,))
valuesNP2  = np.zeros((size2,))
for i in range(size2):
    valuesNP1[i] = pvsDoc1[str(i)]['dbrs'].ndarray('values')[0];
    valuesNP2[i] = pvsDoc2[str(i)]['dbrs'].ndarray('values')[0];
    
im1 = valuesNP1.reshape(size1, size1);    
sub1 = plt.subplot(1,2,1)
plt.imshow(im1)
sub1.set_title(start1)

im2 = valuesNP2.reshape(size1, size1);    
sub2 = plt.subplot(1,2,2)
plt.imshow(im2)
sub2.set_title(start2)   
```

or combine them in one array

```python
# structured array: name, type, count, size
size = pvsDoc1['size'];
pvsNP = np.zeros((size,),dtype=[('pv1', 'a50'), ('value1', 'a10'), ('ts1', 'a20'), ('pv2', 'a50'), ('value2', 'a10'), ('ts2', 'a20') ])
for i in range(size):
    values1 = pvsDoc1[str(i)]['dbrs'].ndarray('values')
    ts1     = pvsDoc1[str(i)]['dbrs'].datetimes('secs', 'nanos')
    values2 = pvsDoc2[str(i)]['dbrs'].ndarray('values')
    ts2     = pvsDoc2[str(i)]['dbrs'].datetimes('secs', 'nanos')
    pvsNP[i] = (pvsDoc1[str(i)]['name'], values1[0], ts1[0], pvsDoc2[str(i)]['name'], values2[0], ts2[0])
```

and print it as pandas-based data frames. 

```python
pd.options.display.max_rows = size 
pvsDF = pd.DataFrame(pvsNP);
pvsDF
```

Finally, we close this session ...

```python
session.close();
```
